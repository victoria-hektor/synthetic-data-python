## **Step 1: Generate Synthetic Data**

```python
import pandas as pd
import numpy as np

# Define the size of the dataset
num_samples = 500

# Generate random synthetic data
np.random.seed(42)  # For reproducibility
synthetic_data = {
    'Feature1': np.random.normal(50, 10, num_samples),  # Normally distributed data
    'Feature2': np.random.uniform(10, 100, num_samples),  # Uniformly distributed data
    'Feature3': np.random.exponential(20, num_samples),  # Exponentially distributed data
    'Category': np.random.choice(['A', 'B', 'C', 'D'], num_samples)  # Categorical data
}

# Convert to a Pandas DataFrame
df = pd.DataFrame(synthetic_data)
print(df.head())
```

## **Step 2: Save Data to a Spreadsheet**

**Save the data as an Excel file for storage and sharing.**

```python
# Save the synthetic data to an Excel file
df.to_excel('synthetic_data.xlsx', index=False)

print("Data saved to 'synthetic_data.xlsx'")
```

## **Step 3: Visualise the Data**
**Histogram** -
Visualise the distribution of numerical features.

```python
import matplotlib.pyplot as plt

# Histogram for Feature1
plt.hist(df['Feature1'], bins=20, alpha=0.7, edgecolor='black')
plt.title('Feature1 Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

**Scatter Plot** - Visualise relationships between two features.

```python
# Scatter plot for Feature1 vs Feature2
plt.scatter(df['Feature1'], df['Feature2'], alpha=0.7)
plt.title('Feature1 vs Feature2')
plt.xlabel('Feature1')
plt.ylabel('Feature2')
plt.show()
```

**Box Plot** - Analyse the spread and outliers in a numerical feature.

```python
# Box plot for Feature3
plt.boxplot(df['Feature3'], vert=False)
plt.title('Feature3 Spread and Outliers')
plt.xlabel('Value')
plt.show()
```

**Bar Plot** - Visualise counts of categorical data.

```python
# Bar plot for Category counts
category_counts = df['Category'].value_counts()
category_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Category Distribution')
plt.xlabel('Category')
plt.ylabel('Frequency')
plt.show()
```

## **Step 4: Data Analysis**

**You can compute basic statistics or perform exploratory analysis.**

```python
# Summary statistics
print(df.describe())

# Count unique categories
print(df['Category'].value_counts())
```
